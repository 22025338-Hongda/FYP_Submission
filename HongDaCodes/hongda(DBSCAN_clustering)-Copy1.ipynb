```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Load the CSV file with the correct encoding
df = pd.read_excel("UpdatedStudentSurvey.xlsx")

# Extracting the relevant features
features = ['Current GPA', 'Absences']
data = df[features].dropna()

# Normalize the data
scaler = StandardScaler()
normalized_data = scaler.fit_transform(data)

# Perform DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(normalized_data)

# Adding the cluster labels to the data
data['Cluster'] = clusters

# Calculate the silhouette score if more than one cluster is found
if len(set(clusters)) > 1:
    score = silhouette_score(normalized_data, clusters)
    print(f'Silhouette Score: {score}')
else:
    print('Silhouette Score: Not applicable (only one cluster or noise)')

# Plotting the scatter plot for clustering
plt.figure(figsize=(10, 6))
plt.scatter(data['Absences'], data['Current GPA'], c=data['Cluster'])
plt.title('DBSCAN Clustering for Current GPA, Absences')
plt.xlabel('Absences')
plt.ylabel('Current GPA')

plt.show()
```

```python
from sklearn.cluster import MeanShift

# Extracting the relevant features
features = ['Age', 'Current GPA']
data = df[features].dropna()

# Normalize the data
scaler = StandardScaler()
normalized_data = scaler.fit_transform(data)

# Perform Mean Shift clustering
mean_shift = MeanShift()
clusters = mean_shift.fit_predict(normalized_data)

# Adding the cluster labels to the data
data['Cluster'] = clusters

# Calculate the silhouette score if more than one cluster is found
if len(set(clusters)) > 1:
    score = silhouette_score(normalized_data, clusters)
    print(f'Silhouette Score: {score}')
else:
    print('Silhouette Score: Not applicable (only one cluster or noise)')

# Plotting the scatter plot for clustering
plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['Current GPA'], c=data['Cluster'], cmap='viridis', marker='o')
plt.title('Mean Shift Clustering for Age and Current GPA')
plt.xlabel('Age')
plt.ylabel('Current GPA')

plt.show()
```

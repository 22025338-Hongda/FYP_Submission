```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("UpdatedStudentSurvey.xlsx")
df
```

```python

avg_gpa = df.groupby('School Satisfaction Level')['Current GPA'].mean().reset_index()
print("Average GPA for each Satisfaction Level group:")
print(avg_gpa)


filtered_avg_gpa = avg_gpa[avg_gpa['Current GPA'] > 3.00]
print("Filtered Average GPA (higher than 3.00):")
print(filtered_avg_gpa)


sorted_filtered_avg_gpa = filtered_avg_gpa.sort_values(by='Current GPA', ascending=False)
print("Sorted Filtered Average GPA (in descending order):")
print(sorted_filtered_avg_gpa)

if not filtered_avg_gpa.empty:
    plt.figure(figsize=(10, 6))
    plt.scatter(sorted_filtered_avg_gpa['School Satisfaction Level'], sorted_filtered_avg_gpa['Current GPA'], color='blue')
    plt.title('Average GPA by Satisfaction Level')
    plt.xlabel('School Satisfaction Level')
    plt.ylabel('Average GPA')
    plt.show()
else:
    print("No Satisfaction Level group with an average GPA higher than 3.00.")
```

```python
avg_gpa = df.groupby('Part-Time Hours')['Current GPA'].mean().reset_index()
print("Average GPA for each Part-Time Hours group:")
print(avg_gpa)

filtered_avg_gpa = avg_gpa[avg_gpa['Current GPA'] > 3.00]
print("Filtered Average GPA (higher than 3.00):")
print(filtered_avg_gpa)

sorted_filtered_avg_gpa = filtered_avg_gpa.sort_values(by='Current GPA', ascending=False)
print("Sorted Filtered Average GPA (in descending order):")
print(sorted_filtered_avg_gpa)

if not filtered_avg_gpa.empty:
    sorted_filtered_avg_gpa = filtered_avg_gpa.sort_values(by='Current GPA', ascending=False)
    plt.figure(figsize=(10, 6))
    plt.scatter(sorted_filtered_avg_gpa['Part-Time Hours'], sorted_filtered_avg_gpa['Current GPA'], color='blue')
    plt.title('Average GPA by Part-Time Hours')
    plt.xlabel('Part-Time Hours')
    plt.ylabel('Average GPA')
    plt.show()
else:
    print("No Part-Time Hours with an average GPA higher than 3.00.")
```

```python
df['No. Of CCAs Participated'] = pd.to_numeric(df['No. Of CCAs Participated'], errors='coerce')

df['Participated in CCA'] = df['No. Of CCAs Participated'] > 0


avg_gpa = df.groupby('Participated in CCA')['Current GPA'].mean().reset_index()
avg_gpa['Participated in CCA'] = avg_gpa['Participated in CCA'].map({True: 'Yes', False: 'No'})
print("Average GPA for participation in CCA:")
print(avg_gpa)


plt.figure(figsize=(10, 6))
sns.barplot(x='Participated in CCA', y='Current GPA', data=avg_gpa, palette='viridis')
plt.title('Average GPA by Participation in CCA')
plt.xlabel('Participated in CCA')
plt.ylabel('Average GPA')
plt.show()
```

```python

df['Game Timing'] = df['Game Timing'].replace("Don't Play Video Games", 0)
df['Game Timing'] = df['Game Timing'].replace("There is not fixed time", 0)
df['Game Timing'] = df['Game Timing'].replace("21 00", 2100)

df['Game Timing'] = df['Game Timing'].astype(int)


avg_gpa = df.groupby('Game Timing')['Current GPA'].mean().reset_index()


plt.figure(figsize=(10, 6))
plt.bar(avg_gpa['Game Timing'], avg_gpa['Current GPA'], alpha=0.5)
plt.title('Average GPA by Game Timing')
plt.xlabel('Game Timing')
plt.ylabel('Average GPA')
plt.xticks(range(0, 2400, 100), rotation=45)
plt.tight_layout()
plt.show()
```
